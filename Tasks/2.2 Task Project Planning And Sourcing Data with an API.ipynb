Step 5. Code Explanation.

```python
# Used this code for the following reasons:
# 1st : Loads pandas library for data analysis
# 2nd : Loads the glob library, which is used for pattern-matching file paths.
# 3rd :  filepaths = glob.glob('../Data/*.csv')
#	glob.glob() searches for all files matching the pattern.
#	Here: "../Data/*.csv" means:
#	Go up one directory (.. from Tasks).
#	Enter the Data folder.
#	Match all files ending in .csv
# 4th : This is the key line:
    #A.	(pd.read_csv(f, encoding="latin1") for f in filepaths)
    #A generator that reads each CSV file into a temporary pandas DataFrame.
    #encoding="latin1" tells pandas how to interpret the raw bytes in the file as text.
    # That’s why I got a UnicodeDecodeError earlier.
    # Using latin1 (ISO-8859-1) is more permissive and can decode any byte 0–255, so it avoids that crash.
    #B.	pd.concat(..., ignore_index=True)
    #Concatenates all the DataFrames from the generator into one big DataFrame.
    #gnore_index=True resets the row index, so you get a clean 0…N index instead of keeping each file’s original row numbers.
```

Basically what we did with this code we took all the csv files and asked to read all the files including the weird symbols so it doesnt get confused. And we combined it in a bigger Data frame.

```python
##Checking all the csv have the same columnd


filepaths = glob.glob('../Data/*.csv')

columns_per_file = {}
for f in filepaths:
    df_temp = pd.read_csv(f, encoding="latin1", nrows=5)  # only peek at first 5 rows
    columns_per_file[f] = df_temp.columns.tolist()

for f, cols in columns_per_file.items():
    print(f"\n{f}:")
    print(cols)
```

```python
#because this is a classic trap when concatenating many CSVs from different months. Sometimes CitiBike (or any dataset provider)
#adds or removes columns between months. If we don’t handle that, pandas will either fail or silently misalign data.
```

```python
#"Always match by the column title. If something is missing on one sheet, just leave it blank.""

filepaths = glob.glob('../Data/*.csv')

dfs = []
for f in filepaths:
    df_temp = pd.read_csv(f, encoding="latin1")
    dfs.append(df_temp)

# Concatenate all, aligning by column names
df = pd.concat(dfs, ignore_index=True, sort=True)
```

What this might mean
- Data quality issue
- Some trips were logged with an end_station_name that isn’t valid (like "NULL", "0", "test station").

- Example: "Grove St PATH" vs "Grove St. PATH " (extra space, dot, casing).
- Pandas treats these as different stations, even though they’re the same.
- System changes
- CitiBike sometimes adds/removes stations, but 285 missing starts is too many to be real.

```python
# 2. Likely causes of mismatch
#	1.	Whitespace and tabs
#	'W 34 St &\\tHudson Blvd E' has a tab \t in the string → counts as a different station.
#	2.	Case differences
#	'William St & Pine St' vs 'william st & pine st' → pandas sees them as different.
#	3.	Trailing/leading spaces
#	'Broad St & Bridge St ' vs 'Broad St & Bridge St' → counts as different.
#	4.	Actual missing stations
#	Some stations may have only existed in certain months → the one “only in START” might be real.
```

```python
# Normalizing stations names
def clean_station_name(s):
    if pd.isna(s):
        return s
    return s.strip().replace("\t", " ").lower()

df["start_station_name"] = df["start_station_name"].apply(clean_station_name)
df["end_station_name"] = df["end_station_name"].apply(clean_station_name)
```

```python
# checking mismatches
start_stations = set(df["start_station_name"].dropna().unique())
end_stations = set(df["end_station_name"].dropna().unique())

only_in_end = end_stations - start_stations
only_in_start = start_stations - end_stations

print("Stations only in END:", len(only_in_end))
print("Stations only in START:", len(only_in_start))
```

```python
# nothing changes so we arrived to the conclusion that is not a cleaning problem, it is more like how data is shown and behaves.
# probably some stations are used only as destinations.
# maybe the data set can be incomplete for certain months.
# probably some stations were temporarely open.
```

```python
# Flaging stations for further analysis if necesary later.
start_stations = set(df["start_station_name"].dropna().unique())
end_stations = set(df["end_station_name"].dropna().unique())

station_types = {}
for s in start_stations.union(end_stations):
    if s in start_stations and s in end_stations:
        station_types[s] = "both"
    elif s in start_stations:
        station_types[s] = "start_only"
    else:
        station_types[s] = "end_only"

```

```python
# checking how many stations fall in each category
from collections import Counter
Counter(station_types.values())
```

- 285 stations are end_only : trips end here but never start.
- 86 stations are both : appear as both start and end.
- 1 station is start_only : trips start here but never end.

For flow analysis or mapping trips, 285 stations are always destinations, 1 is always a start, and 86 are both.

```python
#Load api response as Json


d = json.loads(r.text)
```

```python
d
```

```python
#Converting as wide format for easier analysis

# Convert results to DataFrame
df_weather = pd.DataFrame(d['results'])

# Pivot: one row per date, columns = datatypes
df_weather = df_weather.pivot(index='date', columns='datatype', values='value').reset_index()

# Convert date to datetime
df_weather['date'] = pd.to_datetime(df_weather['date'])

# Preview
print(df_weather.head())
```

```python
# Change of scale values due to NOAA sometimes storing temperatures as tenths of °C and precipitation as hundredths of mm.
df_weather['PRCP'] = df_weather['PRCP'] / 100
df_weather['TAVG'] = df_weather['TAVG'] / 10
df_weather['TMAX'] = df_weather['TMAX'] / 10
df_weather['TMIN'] = df_weather['TMIN'] / 10

df_weather.head()
```

```python
#Ready to merge
# Convert started_at to datetime
df['started_at'] = pd.to_datetime(df['started_at'])

# Create a date-only column for merging
df['date'] = df['started_at'].dt.date

```

```python
# Make sure weather date is also datetime.date
df_weather['date'] = pd.to_datetime(df_weather['date']).dt.date

# Merge CitiBike trips with weather data
df_merged = df.merge(df_weather, on='date', how='left')
df_merged.head()
```

```python
#Explanation:
#	started_at is timestamp with date + time. Weather data is daily.
#	We extract only the date from started_at to align with weather dates.
#	Merge on the new date column using how='left' to keep all CitiBike trips and attach weather info.
#   This will correctly attach the precipitation, average temperature, min/max temp, to each trip.
```

The main reasons for this are likely:
	1.	Date format mismatch – your CitiBike date column might include time, or be in a slightly different format than the weather date.
	2.	Weather data only covers part of the year – your API request returned 1460 rows (about 4 years?), but maybe you’re merging with trips outside those dates.
	3.	Station data differences – if any trip dates fall outside the weather dataset range, they won’t match.
